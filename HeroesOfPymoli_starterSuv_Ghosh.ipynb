## Player Count

```python
#Player count getting unique names
#Purchase data contains non-unique rows, so getting unique number of 'SN'
player_count = len(purchase_data.SN.unique())
print("Total Player count = ", player_count)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate number of unique Items
uniq_items = purchase_data['Item ID'].nunique()
print("Total Unique items = ", uniq_items )

#calculate average price
avg_price = purchase_data['Price'].mean()
#Printing after rounding the value to two place
print("Average Price = ", round(avg_price,2))

#calculate total Number of purchases
tot_purch = len(purchase_data)
print("Total no. of purchase = ", tot_purch)

#calculate total revenue
#Total of price column
tot_rev = purchase_data['Price'].sum()
print("Total Revenue = ", "${:.2f}".format(tot_rev))

#creating dataframe to put all the values
pur_analys = pd.DataFrame({
    "Number of Unique Items": [uniq_items],
    "Average Price" :         ["${:.2f}".format(avg_price)],
    "Number of Purchases":    [tot_purch],
    "Total Revenue":          ["${:.2f}".format(tot_rev)]
    
})

pur_analys
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#calculate total male, female and others
#get_uniq = purchase_data.drop_duplicates(subset="SN",inplace=True)
#get SN and Gender and store in another data frame

# copy earlier purchase data with SN, gender and age to another data frame
#get_uniq = purchase_data[["SN", "Gender", "Age"]] #If I use this Im getting warning message

get_uniq = purchase_data[["SN", "Gender", "Age"]].copy()

#get unique rows, clean up get_uniq, remove duplicate SN
get_uniq.drop_duplicates(subset = "SN", inplace = True)

#find total records where Gender = Male
Male = get_uniq.loc[get_uniq["Gender"] == "Male", :]

#calculate male percent (player count was calculated before)
tot_male_prcnt = (len(Male)/ player_count) * 100

#--------------------------------------------------
#calculate female

#find total records where Gender = Female
Female = get_uniq.loc[get_uniq["Gender"] == "Female", :]

#calculate female percent (player count was calculated before)
tot_female_prcnt = (len(Female)/ player_count) * 100

#--------------------------------------------------
#calculate others = Other / Non-Disclosed

#find total records where Gender = Female
Others = get_uniq.loc[get_uniq["Gender"] == "Other / Non-Disclosed", :]

#calculate other percent (player count was calculated before)
tot_other_prcnt = (len(Others)/ player_count) * 100

#--------------------------------------------------------

#creating dataframe to put all the new values:
gender_dm = pd.DataFrame({
   # "Gender": ["Male", "Female", "Other / Non-Disclosed"],
   # index =["Male", "Female", "Other / Non-Disclosed"],
    "Total Count" : [len(Male), len(Female), len(Others)],
    "Percentage of Players":    [tot_male_prcnt, tot_female_prcnt, tot_other_prcnt]
})

#change the index name to male, female, others
gender_dm = pd.DataFrame.from_records(gender_dm, index =["Male", "Female", "Other / Non-Disclosed"])

#formating the Percentage column 

gender_dm["Percentage of Players"] = gender_dm["Percentage of Players"].map("{:.2f}".format)

gender_dm
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#calculate total male purchase(include non-unique):
#pick only records gender = male
male_data = purchase_data.loc[purchase_data["Gender"] == "Male", :]

m_pur_count = len(male_data) #Male purchase count
m_tot_pur = male_data["Price"].sum() # total price of male purchases (total price column)
m_avg_price = m_tot_pur/m_pur_count # average male purchase price
m_avg_person = m_tot_pur/len(Male)#average male purchase 

#calculate total female purchase
female_data = purchase_data.loc[purchase_data["Gender"] == "Female", :]

f_pur_count = len(female_data)
f_tot_pur = female_data["Price"].sum()
f_avg_price = f_tot_pur/f_pur_count
f_avg_person = f_tot_pur/len(Female)

#calculate total Others purchase
other_data = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]

o_pur_count = len(other_data)
o_tot_pur = other_data["Price"].sum()
o_avg_price = o_tot_pur/o_pur_count
o_avg_person = o_tot_pur/len(Others)


#creating dataframe to put all the values
#Gender Analytics
gender_ana = pd.DataFrame({
        "Gender": ["Male", "Female", "Other / Non-Disclosed"],
         "Purchase Count" : [m_pur_count, f_pur_count, o_pur_count],
    "Average Purchase Price": [m_avg_price, f_avg_price, o_avg_price],
    "Total Purchase Value": [m_tot_pur, f_tot_pur, o_tot_pur],
    "Avg Total Purchase per Person": [m_avg_person, f_avg_person, o_avg_person]
    
})

#set index by gender
gender_ana.set_index("Gender", inplace = True)

#sort index by Gender
gender_ana_sort = gender_ana.sort_values("Gender")

#formating the columns
gender_ana_sort["Average Purchase Price"] = gender_ana_sort["Average Purchase Price"].map("${:.2f}".format)
gender_ana_sort["Total Purchase Value"] = gender_ana_sort["Total Purchase Value"].map("${:.2f}".format)
gender_ana_sort["Avg Total Purchase per Person"] = gender_ana_sort["Avg Total Purchase per Person"].map("${:.2f}".format)

#pur_age_demo["Total Purchase Value"] = pur_age_demo["Total Purchase Value"].map("${:.2f}".format) 

gender_ana_sort
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

Itemid= purchase_data.groupby(["Item ID","Item Name"])

I_count = Itemid.Price.count()
#average_purchase_price = Price.mean()
iprice = Itemid.Price.mean()

total_purval = Itemid.Price.sum()

itemdf = pd.DataFrame({
    "Purchase Count":I_count, 
    "Item Price":iprice,
    "Total Purchase Value":total_purval
    })

#sort on Purchase count
item_sort = itemdf.sort_values(["Purchase Count"], ascending = False)

item_sort["Item Price"] = item_sort["Item Price"].map("${:.2f}".format)
item_sort["Total Purchase Value"] = item_sort["Total Purchase Value"].map("${:.2f}".format)

item_sort.head()
```
